```python
# Weighted global mean for Average Rating
mu_rating = (
    filtered_df["Average Rating"] * filtered_df["Number of ratings"]
).sum() / filtered_df["Number of ratings"].sum()

# Weighted global mean for Average Difficulty
mu_difficulty = (
    filtered_df["Average Difficulty"] * filtered_df["Number of ratings"]
).sum() / filtered_df["Number of ratings"].sum()

print("Weighted global mean rating:", round(mu_rating, 3))
print("Weighted global mean difficulty:", round(mu_difficulty, 3))
```

```python
k = 5

# Bayesian-adjusted Average Rating
filtered_df["AvgRating_Bayes"] = (
    filtered_df["Average Rating"] * filtered_df["Number of ratings"]
    + mu_rating * k
) / (filtered_df["Number of ratings"] + k)

# Bayesian-adjusted Average Difficulty
filtered_df["AvgDifficulty_Bayes"] = (
    filtered_df["Average Difficulty"] * filtered_df["Number of ratings"]
    + mu_difficulty * k
) / (filtered_df["Number of ratings"] + k)

# Inspect
filtered_df[
    ["Average Rating", "AvgRating_Bayes",
     "Average Difficulty", "AvgDifficulty_Bayes",
     "Number of ratings"]
].head(10)
```

```python
import numpy as np
from scipy.stats import ks_2samp, levene

# Split ratings by gender
male_ratings = filtered_df.loc[filtered_df['Gender'] == 0, 'AvgRating_Bayes']
female_ratings = filtered_df.loc[filtered_df['Gender'] == 1, 'AvgRating_Bayes']

alpha = 0.005

# -----------------------------
# 1️⃣ Kolmogorov-Smirnov test
# -----------------------------
ks_stat, ks_p = ks_2samp(male_ratings, female_ratings)
print("Kolmogorov-Smirnov Test (male vs female)")
print(f"KS statistic: {ks_stat:.4f}")
print(f"p-value: {ks_p:.4f}")
if ks_p < alpha:
    print(f"Result is significant at alpha={alpha}: distributions differ")
else:
    print(f"Result is NOT significant at alpha={alpha}: no strong evidence distributions differ")

# -----------------------------
# 2️⃣ ECDF plot
# -----------------------------
def ecdf(data):
    """Compute x, y values for ECDF"""
    x = np.sort(data)
    y = np.arange(1, len(data)+1) / len(data)
    return x, y

x_male, y_male = ecdf(male_ratings)
x_female, y_female = ecdf(female_ratings)

plt.figure(figsize=(6,4))
plt.step(x_male, y_male, label='Male', color='skyblue')
plt.step(x_female, y_female, label='Female', color='lightpink')
plt.xlabel('AvgRating_Bayes')
plt.ylabel('ECDF')
plt.title('ECDF of Bayesian Average Ratings by Gender')
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 3️⃣ Levene's test for variance
# -----------------------------
lev_stat, lev_p = levene(male_ratings, female_ratings)
print("\nLevene's Test for Equality of Variances (male vs female)")
print(f"Levene statistic: {lev_stat:.4f}")
print(f"p-value: {lev_p:.4f}")
if lev_p < alpha:
    print(f"Result is significant at alpha={alpha}: variances differ")
else:
    print(f"Result is NOT significant at alpha={alpha}: no strong evidence variances differ")
```

```python
import numpy as np

# Split ratings by gender
male_ratings = filtered_df.loc[filtered_df['Gender'] == 0, 'AvgRating_Bayes'].values
female_ratings = filtered_df.loc[filtered_df['Gender'] == 1, 'AvgRating_Bayes'].values

n_m = len(male_ratings)
n_f = len(female_ratings)
n_boot = 10000
rng = np.random.default_rng(seed=42)

# -----------------------------
# 1️⃣ Effect size for average rating
# -----------------------------
mean_diff = male_ratings.mean() - female_ratings.mean()

# Cohen's d
s_pooled = np.sqrt(((n_m - 1)*male_ratings.std(ddof=1)**2 + (n_f - 1)*female_ratings.std(ddof=1)**2) / (n_m + n_f - 2))
cohens_d = mean_diff / s_pooled

# 95% CI for mean difference using bootstrap
boot_mean_diffs = []
for _ in range(n_boot):
    male_sample = rng.choice(male_ratings, size=n_m, replace=True)
    female_sample = rng.choice(female_ratings, size=n_f, replace=True)
    boot_mean_diffs.append(male_sample.mean() - female_sample.mean())

ci_mean_diff = np.percentile(boot_mean_diffs, [2.5, 97.5])

# -----------------------------
# 2️⃣ Effect size for spread (variance ratio)
# -----------------------------
var_m = np.var(male_ratings, ddof=1)
var_f = np.var(female_ratings, ddof=1)
var_ratio = var_m / var_f  # effect size for spread

# 95% CI for variance ratio using bootstrap
boot_var_ratios = []
for _ in range(n_boot):
    male_sample = rng.choice(male_ratings, size=n_m, replace=True)
    female_sample = rng.choice(female_ratings, size=n_f, replace=True)
    boot_var_ratios.append(np.var(male_sample, ddof=1) / np.var(female_sample, ddof=1))

ci_var_ratio = np.percentile(boot_var_ratios, [2.5, 97.5])

# -----------------------------
# Print results
# -----------------------------
print("Effect size for average rating (male - female):")
print(f"Mean difference: {mean_diff:.4f}")
print(f"Cohen's d: {cohens_d:.4f}")
print(f"95% CI for mean difference: [{ci_mean_diff[0]:.4f}, {ci_mean_diff[1]:.4f}]\n")

print("Effect size for spread (variance ratio male/female):")
print(f"Variance ratio: {var_ratio:.4f}")
print(f"95% CI for variance ratio: [{ci_var_ratio[0]:.4f}, {ci_var_ratio[1]:.4f}]")
```

```python

from scipy.stats import chi2_contingency, fisher_exact

# Define your tag columns explicitly (replace with your actual tag names)
tag_columns = ['Hot or Not', 'Take class again', 'Online class ratings',
               'Tough grader', 'Good feedback', 'Respected', 'Lots to read',
               'Participation matters', 'Don’t skip class or you will not pass',
               'Lots of homework', 'Inspirational', 'Pop quizzes!', 'Accessible',
               'So many papers', 'Clear grading', 'Hilarious', 'Test heavy',
               'Graded by few things', 'Amazing lectures', 'Caring', 
               'Extra credit', 'Group projects', 'Lecture heavy']

# Step 1: Convert all tags to binary
for tag in tag_columns:
    filtered_df[f"{tag}_binary"] = (filtered_df[tag] > 0).astype(int)

binary_tag_columns = [f"{tag}_binary" for tag in tag_columns]

# Step 2: Fisher / Chi-square analysis
results = []

for tag in binary_tag_columns:
    table = pd.crosstab(filtered_df["Gender"], filtered_df[tag])
    
    # Use Fisher if any cell < 5, else chi-square
    if (table.values < 5).any():
        _, p = fisher_exact(table)
    else:
        _, p, _, _ = chi2_contingency(table)
    
    male_prop = filtered_df[filtered_df["Gender"]==0][tag].mean()
    female_prop = filtered_df[filtered_df["Gender"]==1][tag].mean()
    
    results.append({
        "Tag": tag.replace("_binary",""),
        "Male %": male_prop,
        "Female %": female_prop,
        "Difference in average ratings": female_prop - male_prop,
        "p-value": p,
        "Significant": p < 0.005
    })

# Step 3: Convert results to DataFrame and sort
chi_results = pd.DataFrame(results).sort_values("p-value").reset_index(drop=True)
chi_results
```

Predictive performance: The model with Take class again is clearly better, because both R² is much higher and RMSE is much lower on both train and test.

Trade-off: It uses slightly fewer observations, but the gain in explainability and accuracy outweighs the cost of losing a few rows.It looks like the test R² is slightly higher than the train R², which seems counterintuitive because usually models perform slightly worse on unseen data.

Why this can happen

Random train-test split

Your dataset is relatively small or the split is lucky

The test set may, by chance, be easier to predict than the training set.
Here the difference is tiny (0.7991 → 0.8031)

This is normal and not a concern — it does not indicate overfitting or a problem as the RMSE's of both test and train are low, ie there is no sig difference betwneeen RMSEs
Take class again is likely highly predictive of AvgRating_Bayes.

Dropping it:

Model loses an important explanatory variable

R² drops from ~0.80 → ~0.49

Variance of the outcome (y) itself isn’t lower, but the proportion explained by the predictors is lower.

So yes — the lower R² is mostly due to losing an important predictor, not necessarily due to low variance in the dataset.

```python
# Drop AvgRating_Bayes and Take class again
X_cols = [
    'Number of ratings','AvgDifficulty_Bayes','Online class ratings','Gender',
    'Online class ratings_binary','Tough grader_binary',
    'Good feedback_binary','Respected_binary','Lots to read_binary',
    'Participation matters_binary',
    'Don’t skip class or you will not pass_binary',
    'Lots of homework_binary','Inspirational_binary',
    'Pop quizzes!_binary','Accessible_binary','So many papers_binary',
    'Clear grading_binary','Hilarious_binary','Test heavy_binary',
    'Graded by few things_binary','Amazing lectures_binary',
    'Caring_binary','Extra credit_binary','Group projects_binary',
    'Lecture heavy_binary'
]

Y_col = 'Hot or Not'

# Drop rows with NaNs in these columns
clf_df = filtered_df[X_cols + [Y_col]].dropna().copy()

X = clf_df[X_cols]
y = clf_df[Y_col]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix



X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

logit_pipeline = Pipeline([
    ('scaler', StandardScaler()),   # scale numeric features
    ('logit', LogisticRegression(
        penalty='l2',
        class_weight='balanced',   # handle class imbalance
        max_iter=1000,
        solver='lbfgs'
    ))
])

logit_pipeline.fit(X_train, y_train)
```

```python
y_train_pred = logit_pipeline.predict(X_train)
y_test_pred = logit_pipeline.predict(X_test)

y_train_prob = logit_pipeline.predict_proba(X_train)[:, 1]
y_test_prob = logit_pipeline.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import f1_score

# AUROC
auc_train = roc_auc_score(y_train, y_train_prob)
auc_test = roc_auc_score(y_test, y_test_prob)

# Accuracy
acc_train = accuracy_score(y_train, y_train_pred)
acc_test = accuracy_score(y_test, y_test_pred)

# F1 scores
f1_train = f1_score(y_train, y_train_pred)
f1_test = f1_score(y_test, y_test_pred)

print("Logistic Regression (Pepper Prediction, dropped AvgRating_Bayes & Take class again)")
print(f"Train AUROC: {auc_train:.3f}, Accuracy: {acc_train:.3f}, F1: {f1_train:.3f}")
print(f"Test  AUROC: {auc_test:.3f}, Accuracy: {acc_test:.3f}, F1: {f1_test:.3f}")

# Confusion matrix (Test)
print("\nConfusion Matrix (Test):")
print(confusion_matrix(y_test, y_test_pred))
```
